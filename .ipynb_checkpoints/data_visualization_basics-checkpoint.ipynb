```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100000)
plt.plot(x, np.sin(x))
```

```python
plt.scatter(x, np.sin(x))
```

```python
plt.scatter(x[::1000], np.sin(x)[::1000])
```

```python
plt.style.use('classic')
x = np.linspace(0, 100, 1000)
y = np.cumsum(np.random.randn(1000, 5), 0)
plt.plot(x, y)
plt.legend('ABCDE', ncol=2, loc='upper left')
plt.show();
```

```python
import seaborn as sns
sns.set()
plt.plot(x, y)
plt.legend('ABCDE', ncol=2, loc='upper left')
plt.show()
```

## Histograms and KDE plots

```python
import pandas as pd
sns.set()
X = pd.Series(50*np.random.randn(5000))
Y = pd.Series(200*np.random.randn(5000))
Z = pd.Series(100*np.random.randn(5000)+500)
data = pd.DataFrame({'X':X, 'Y':Y, 'Z':Z})
for col in data.columns:
    plt.hist(data[col], density=True, alpha=.5)
```

```python
for col in data.columns:
    sns.displot(data[col])
```

```python
for col in data.columns:
    sns.histplot(data[col])
```

```python
# future warning
for col in data.columns:
    sns.distplot(data[col])
```

## Pairplot and Jointplot

```python
with sns.axes_style('whitegrid'):
    sns.jointplot(x='X', y='Y', data=data, kind='kde')
```

```python
sns.pairplot(data)
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
```
